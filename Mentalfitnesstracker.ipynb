```python
df = pd.DataFrame(data)
df.head(10)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in df.columns:
  if df[i].dtype =='object':
    df[i]=l.fit_transform(df[i])
```

```python
df.shape
```

Split Data = (6840, 10)
in this step, we are going to split data in to two parts (training and testing), so thatwe can train our model on training dataset and test its accuracy on unseen test data

```python
x= df.drop('mental_fitness', axis =1)
y= df['mental_fitness']
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest=train_test_split(x,y,test_size=.20, random_state=2)
```

```python
print('xtrain',xtrain.shape)
print('xtest',xtest.shape)
print('ytrain',ytrain.shape)
print('ytest',ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
lr =LinearRegression()
lr.fit(xtrain,ytrain)

ytrain_pred =lr.predict(xtrain)
mse =mean_squared_error(ytrain,ytrain_pred)
rmse= (np.sqrt(mean_squared_error(ytrain, ytrain_pred)))

r2=r2_score(ytrain, ytrain_pred)

print("The Linear Regression Model P erformance for trainig set")
print(".............")

print('mse is {}'.format(mse))
print('Rmse is {}'.format(rmse))
print('R2 score is  {}'.format(r2))


```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(xtrain,ytrain)

ytrain_pred =rf.predict(xtrain)
mse =mean_squared_error(ytrain,ytrain_pred)
rmse= (np.sqrt(mean_squared_error(ytrain, ytrain_pred)))

r2=r2_score(ytrain, ytrain_pred)

print("The RandomForestRegressor Model Performance for trainig set")
print(".............")

print('mse is {}'.format(mse))
print('Rmse is {}'.format(rmse))
print('R2 score is  {}'.format(r2))
```

Evaluation:
In this we will  compare the values of above two models


```python
# Linea r regression model evaluation for testing set
ytest_pred =lr.predict(xtest)
mse = mean_squared_error(ytest,ytest_pred)
rase =(np.sqrt(mean_squared_error(ytest, ytest_pred)))
r2= r2_score(ytest,ytest_pred)

print("The  linear resgression model performance for testing set")
print("..........")
print('mse is {}'.format(mse))
print('rmse is {}'.format(rmse))
print('R2 score is {}'.format(r2))


ytest_pred=rf.predict(xtest)
mse =mean_squared_error(ytest, ytest_pred)
rmse= (np.sqrt(mean_squared_error(ytest, ytest_pred)))
r2 = r2_score(ytest, ytest_pred)

```

```python
#RandomForestRegressor model wvaluation for testing set
ytest_pred =rf.predict(xtest)
mse = mean_squared_error(ytest,ytest_pred)
rase =(np.sqrt(mean_squared_error(ytest, ytest_pred)))
r2= r2_score(ytest,ytest_pred)

print("\n \nThe  Random forest resgression model performance for testing set")
print("..........")
print('mse is {}'.format(mse))
print('rmse is {}'.format(rmse))
print('R2 score is {}'.format(r2))


```
